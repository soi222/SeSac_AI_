```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## sorting

```python
dict_data = {
    "c0":np.arange(1,4),
    "c1":np.arange(4,7),
    "c2":np.arange(7,10),
    "c3":np.arange(10,13),
    "c4":np.arange(13,16)
}

dic_df = pd.DataFrame(dict_data, index = ["r0","r1","r2"])
print(dic_df)
```

```python
## index 정렬, Transfer
dic_df.sort_index(ascending= False, inplace= True)
print("df = \n", dic_df)

## value
print("\n", "*"*100)
dic_df.sort_values(by = "c0", ascending= True, inplace= True) #ascending :오름차순
print("df = \n", dic_df)
```

```python
#선생님의 답
import pandas as pd
fname = "vor_r.xlsx"

vor = pd.read_excel('./dataset/vor_r.xlsx')
print("vor = \n",vor.head())

vor_data.rename(columns = {"group":"소속","id":"번호"},inplace = True)
print()
print("vor = \n",vor.head())

print("vor = \n", vor)
print("\n", "*"*100)

mean_col = vor.loc[:,"hz0.04":].mean(axis= 1) 

vor.insert(3, "mean", mean_col)
vor.sort_values(by = "mean", ascending=False, inplace= True)
vor_df10 = vor.iloc[:10,:4]
```

```python
## Filter

vor_df10.iloc[5:,2] = "Post"
vor_df10
```

```python
idx1 = vor_df10.loc[:,"time"] != "Pre"
print("time != pre \n",idx1)
print("\n", "*"*100)
idx2 = vor_df10.loc[:,"mean"] > 0.77
print("mean > 0.77 \n",idx2)

print("\n", "*"*100)
print(vor_df10[(idx1) & (idx2) & (vor_df10.loc[:,"group"] == 1)]) #and
print(vor_df10[(idx1) | (idx2)]) #or
```
